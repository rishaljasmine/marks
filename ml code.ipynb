```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"5AI REGULAR.xls"
df = pd.read_excel(file_path, header=4)

column1 = "Internal.3"
column2 = "External.3"
selected_data = df[[column1, column2]]

plt.figure(figsize=(8, 6))
plt.scatter(selected_data[column1], selected_data[column2], color="blue", alpha=0.7)
plt.title(f"Scatter Plot of {column1} vs {column2}")
plt.xlabel(column1)
plt.ylabel(column2)
plt.grid(True)
plt.show()

correlation_coefficient = np.corrcoef(selected_data[column1], selected_data[column2])[0, 1]
print(f"Pearson Correlation Coefficient ({column1} vs {column2}): {correlation_coefficient:.2f}")

cov_matrix = np.cov(selected_data[column1], selected_data[column2])
print("\nCovariance Matrix:")
print(cov_matrix)

correlation_matrix = selected_data.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = r"5AI REGULAR.xls"
df = pd.read_excel(file_path, header=4)


internal_cols = [col for col in df.columns if 'Internal' in col]
external_cols = [col for col in df.columns if 'External' in col]


df['Total_Internal'] = df[internal_cols].sum(axis=1)
df['Total_External'] = df[external_cols].sum(axis=1)


plt.figure(figsize=(8, 6))
plt.scatter(df['Total_Internal'], df['Total_External'], color="green", alpha=0.7)
plt.title("Scatter Plot of Total Internal vs Total External Marks")
plt.xlabel("Total Internal Marks")
plt.ylabel("Total External Marks")
plt.grid(True)
plt.show()


correlation_coefficient = np.corrcoef(df['Total_Internal'], df['Total_External'])[0, 1]
print(f"Pearson Correlation Coefficient (Total Internal vs Total External): {correlation_coefficient:.2f}")


cov_matrix = np.cov(df['Total_Internal'], df['Total_External'])
print("\nCovariance Matrix:")
print(cov_matrix)


correlation_matrix = df[['Total_Internal', 'Total_External']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Correlation Matrix Heatmap (Total Marks)")
plt.show()
```
